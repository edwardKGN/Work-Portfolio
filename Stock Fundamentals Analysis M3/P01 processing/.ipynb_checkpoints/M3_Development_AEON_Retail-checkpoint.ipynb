# Concept

New approach is to use df.apply(func) to extract financial data and process it for Financial Ratios calculation for better computation efficiency (lower computation load, higher speed).

Financial ratios are split into numerator and denominator of different financial data sets in general.

```python
import pandas as pd
from pandas import Series, DataFrame

import numpy as np
```

```python
df_input = pd.read_excel("uniqueItemsFromAnnualReports_AEON_Retail_M3.xlsx", sheet_name = "Main", index_col = 0)
```

```python
# df_input.drop("Source Document", axis = 1, inplace = True)
```

```python
df_input
```

```python
financialYears = df_input.columns.values
```

```python
import itertools

column_names =  list(itertools.chain(["Financial Ratios"], financialYears))
```

```python
df_financialRatios = pd.DataFrame(columns = column_names)
df_financialRatios
```

```python
# Read Row for Financial Ratios calculation
def read_row(row):
    print(row, type(row))
    print(row.name)
    print(row.index)
    print(row.values)
    
    return row
```

```python
df_input_row = df_input.loc[["Unit Price (Currency)"]].apply(read_row, axis = 1)
```

```python
df_input_row
```

```python
df_input.loc[["Unit Price (Currency)", "Outstanding Shares"]]
```

```python
ser_processing = df_input.loc[["Unit Price (Currency)", "Outstanding Shares"]].apply(read_row, axis = 1)
```

```python
# Read Row
arr_unitPrice = df_input.loc["Unit Price (Currency)"].values
```

```python
arr_outstandingShares = df_input.loc["Outstanding Shares"].values
```

```python
arr_marketCap = (arr_unitPrice * arr_outstandingShares) / 1000
```

```python
arr_marketCap 
```

```python
arr_totAssets = df_input.loc["Total Assets"].values
arr_intAssets = df_input.loc["Intangible Assets"].values
arr_totLiabilities = df_input.loc["Total Liabilities"].values

arr_bookVal = arr_totAssets - arr_intAssets - arr_totLiabilities
arr_bookVal
```

```python
arr_PE = arr_marketCap/arr_bookVal
```

```python
df_financialRatios = df_financialRatios.append({"Financial Ratios": "Price to Book Ratio"}, ignore_index = True, sort = False)
```

```python
df_financialRatios
```

```python
# Write to df_FinancialRatios
for index, item in enumerate(arr_PE):
    # print(index, item)
    df_financialRatios.loc[df_financialRatios[df_financialRatios.columns[0]] == "Price to Book Ratio", financialYears[index]] = item

df_financialRatios
```
